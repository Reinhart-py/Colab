Now, we will create a place for you to paste the list of email accounts you want to use.

Format

Email:password

Email:password

Example:

bimsi@gmail.com:bshshsh727
reinhart@gmail.com: heb66#&hsb

```python
from IPython.display import display
import ipywidgets as widgets

# Create a large text area
text_box = widgets.Textarea(
    value='',
    placeholder='Paste your email:password list here...',
    description='Accounts:',
    layout=widgets.Layout(width='100%', height='300px')
)

# Create a Save button
button = widgets.Button(description="Save to accounts.txt", button_style='success')

output = widgets.Output()

def save_to_file(b):
    with open("accounts.txt", "w") as f:
        f.write(text_box.value)
    with output:
        output.clear_output()
        print("✅ Saved to accounts.txt")

button.on_click(save_to_file)

# Show everything
display(text_box, button, output)
```

After you paste your email accounts and click the "Save to accounts.txt" button above, this command will show you what is saved in the file to make sure it's correct.
